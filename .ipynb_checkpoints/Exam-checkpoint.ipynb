#### Dealing with non-numeric columns

As you might already notice, the analyzed dataset is quite big, that is why we try to optimaze the allocated space and time for execution of applied methods. In order to lower the size of the data, we may change its data type. As the calue for the time column does not exceed 255, we may assign them the type "uint8". Notice that the space consumed by the data lowered almost twice (from 4+ GB to 2.4+ GB). 

```python
simplified_columns = {'departure_weekday': np.uint8,
                   'departure_month': np.uint8,
                   'departure_day': np.uint8,
                   'departure_hour': np.uint8,
                   'departure_minute': np.uint8,
                   'arrival_weekday': np.uint8,
                   'arrival_month': np.uint8,
                   'arrival_day': np.uint8,
                   'arrival_hour': np.uint8,
                   'arrival_minute': np.uint8}
df = df.astype(simplified_columns)
df.info()
```

As we have already noticed earlier, in the analyzed dataset there are several columns of type "object". Knowing that prediction models do not like words, our goal now is to transform these columns into numeric data type by replasing each class of the object with an integer.
For this purpose, we are going to use OrdinalEncoder over the columns of type object. Also, we do not want the "NAN" values in "vehicle_class" and "fare" coumns to be interpreted as classes, that is why we do not mind them when replacing classes. 
In order to simplify the code, we are going to write a function that perform this replacing.

```python
# Preparing for encoding and fill the NANs in vehicle_class and fare
cat_cols = ['origin', 'destination', 'vehicle_type', 'vehicle_class', 'fare']
```

```python
# Initiate models
encoder = OrdinalEncoder()
```

```python
# Function to replace not null object values with ordinals
def encode(data):
    non_null = np.array(data.dropna())
    impute_ordinal = encoder.fit_transform(non_null.reshape(-1, 1))
    data.loc[data.notnull()] = np.squeeze(impute_ordinal)
    return data
```

```python
# Apply the function for each column of type "object"
for columns in cat_cols:
    encode(df[columns])
```

As we an observe, the categorical variables were replaced by integers (they are write as floats, tho) and the "NAN" values are on their places.

```python
df.head()
```

Again, we change the data type of the columns in dataset in order to reduce the size of the data. As you may observe, it reduces to 1.2 GB now, that is a quite good improvement that allows us to perform operations a bit faster.

As an alternative method to deal with NAN values we are going to fill the "fare" and "vehicle_class"columns with the mode values. When it comes to "price", we are going to substitute the nulls by the mean of the column grouped by "fare", "vehicle_class", "vehicle_type", "duration" in case it is not null and by the oveall mean otherwise. 
Of course this method is not so effective and it can cause data lackage, but it is still better than dropping columns or create a new class, which may result in a poor performance due to introducing new features into the data.

```python
# a function to fill the nulls with mode
def impute_mode(column, data):
    data[column].fillna(data[column].mode()[0], inplace = True)
```

```python
impute_mode('vehicle_class', df)
impute_mode('fare', df)
df.head()
```

```python
df.isnull().sum()
```

```python
# replace null values in "price" column
over_avg = df.price.mean()
avg = lambda x: x.mean() if x.notna().any() else over_avg
df['price'] = df['price'].fillna(df.groupby(['fare', 'vehicle_class', 'vehicle_type', 'duration']).price.transform(avg))
```

```python
df.isnull().sum()
```

```python
df.head()
```

Recall the fact that we still speak about categorical data. In this case, we aren't really in place to speak about normalized data or stuff like this. A way to overcome this problem is by using a cardinality reducing encoding, meaning replacing the integers of each class by the frequence of the class or the mean of the target for this class. 
In this case, we are going to replace these features by their frequencies. 

```python
freq = FrequencyImputationTransformer()
df_freq = freq.apply(df[['origin', 'destination', 'vehicle_type', 'vehicle_class', 'fare', 'duration']], target = 'duration', columns = ['origin', 'destination', 'vehicle_type', 'vehicle_class', 'fare'])
```

```python
df_freq = df_freq.drop(['duration'], axis = 1)
df = df.drop(['origin', 'destination', 'vehicle_type', 'vehicle_class', 'fare'], axis = 1).join(df_freq)
df
```

## 4. Post-Processing and model building

First of all, let us define our feature set and target. Also, as the target column has continuous values, it means we have to deal with a regression problem. Therefore, let us define the models we are going to use: LinearRegression (as it uses variable dependences) and DecisionTreeRegressor (as it is a tree type regressor, good when dealing with non-linear data).

```python
X = df.drop(['duration'], axis = 1)
y = df['duration']
```

We build a pipeline that consists of StandardScaler scaler, in order to scale the data and bring it in the range from -3 to 3 according to the both axes,  and a model we would like to perfom.

```python
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LinearRegression())
])
```

For the first prediction, let us use all values and see the accuracy of the result. 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
```

```python
print('Mean Absolute Error: ', mean_absolute_error(y_test, y_pred))
```

```python
pipe2 = Pipeline([
    ('scaler', StandardScaler()),
    ('model', DecisionTreeRegressor(max_depth = 2))
])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
pipe2.fit(X_train, y_train)
y_pred2 = pipe2.predict(X_test)
```

```python
print('Mean Absolute Error: ', mean_absolute_error(y_test, y_pred2))
```

Now, let us use a feature selection model and see if it gives a better result. 
Recall that we have a regression problem. In this case, we use PValueSelector in order to select some columns. As you may notice, returns all the columns, meaning all of them are important in achieving a good accuracy of the model.
Also, it was tried to apply LASSO Feature Selector, but got into a memory error as the dataset is quite big. 

```python
pvalue = PValueSelector()
selected_columns = pvalue.select(df, 'duration')
print(selected_columns)
```
